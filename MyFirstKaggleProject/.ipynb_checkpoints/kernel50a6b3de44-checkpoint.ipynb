```python
from sklearn.preprocessing import LabelEncoder

for column in test.columns:
        if test[column].dtype == np.number:
            continue
        test[column] = LabelEncoder().fit_transform(test[column])
for column in train.columns:
        if train[column].dtype == np.number:
            continue
        train[column] = LabelEncoder().fit_transform(train[column])
```

```python
train.head()
```

```python
test.head()
```

```python
test= test.drop('EmployeeNumber', axis = 1)
train = train.drop('EmployeeNumber', axis = 1)

```

```python
test.corr()
```

```python
train.corr()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,20))
sns.heatmap(test.corr(),annot=True)
```

```python
xtest=test
ytrain=train['Attrition']
xtrain=train.drop('Attrition',axis = 1)
```

```python
from sklearn.ensemble import RandomForestClassifier
f = RandomForestClassifier(n_estimators=10,criterion='entropy',random_state=47)
f.fit(xtrain,ytrain)
```

```python
f.score(xtrain,ytrain)
```

```python
now = f.predict_proba(xtest)
```

```python
final = now[:,1]
data=data.drop('Attrition',axis=1)
data['Attrition']=final
data.to_csv('submission.csv',index=False)
```
